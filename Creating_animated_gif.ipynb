```python
#visualize the rotations of the U gates on a Bloch sphere
#use the plot_bloch_multivector() and plot_state_qsphere() methods to visualize how our state vectors behave
#what the possible outcomes are
#These both provide static views of the qubit at one moment and of the U
#gates for a specific set of angles.


from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere

from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
import numpy as np
from PIL import Image
import os
# Import image and file processing tools
from PIL import Image
import os
import numpy as np
print("Animating the U gates")
print("--------------------------")

#  an /images directory at the same location as the script.
dirName = 'images'
print("Checking if /images directory exists...")
if not os.path.exists(dirName):
    os.mkdir(dirName)
    print("Directory" , dirName ,  "created ")
else:    
    print("Directory" , dirName ,  "exists")

# Each circuit is then run through the
# statevector simulator in the get_psi() function 
# Visualizing the quantum gates
def get_psi(circuit):
    global psi
    backend = Aer.get_backend('statevector_simulator') 
    result = execute(circuit, backend).result()
    psi = result.get_statevector(circuit)
    return(psi)


#setting the input parameters and all internal function variables:
def create_images(gate, theta=0.0, phi=0.0, lam=0.0):
    steps = 20.0
    theta_steps = theta / steps
    phi_steps = phi / steps
    lam_steps = lam / steps
    n, theta, phi, lam = 0, 0.0, 0.0, 0.0

    global q_images, b_images, q_filename, b_filename
    b_images = []
    q_images = []
    b_filename = "animated_qubit"
    q_filename = "animated_qsphere"

    while n < steps + 1:
        qc = QuantumCircuit(1)
        if gate == "u3":
          qc.u(theta, phi, lam, 0)
          title = f"U3: θ = {round(theta, 2)} φ = {round(phi, 2)} λ = {round(lam, 2)}"
        elif gate == "u2":
          qc.u(np.pi / 2, phi, lam, 0)
          title = f"U2: φ = {round(phi, 2)} λ = {round(lam, 2)}"
        elif gate == "u1":
          qc.u(0, 0, phi, 0)
          title = f"U1: λ = {round(phi, 2)}"
        else:
          print("Invalid gate.")
          return

        plot_bloch_multivector(get_psi(qc), title).savefig(f'images/bloch{n}.png')
        imb = Image.open(f'images/bloch{n}.png')
        b_images.append(imb)

        plot_state_qsphere(psi).savefig(f'images/qsphere{n}.png')
        imq = Image.open(f'images/qsphere{n}.png')
        q_images.append(imq)

        n += 1
        theta += theta_steps
        phi += phi_steps
        lam += lam_steps

# Create and save the animated GIFs
def save_gif(gate):
    duration=100
    b_images[0].save(gate+'_'+b_filename+'.gif',
               save_all=True,
               append_images=b_images[1:],
               duration=duration,
               loop=0)
    q_images[0].save(gate+'_'+q_filename+'.gif',
               save_all=True,
               append_images=q_images[1:],
               duration=duration,
               loop=0)
    print("Bloch sphere animation saved as: \n"+os.getcwd()+"/"+gate+"_"+b_filename+".gif"+"\nQsphere animation saved as: \n"+os.getcwd()+"/"+gate+"_"+q_filename+".gif")

# Main loop
def main(): 
    global gate
    gate=""
    theta=0.0
    phi=0.0
    lam=0.0
    while gate !="exit": 
        gate=input("Enter u1, u2, or u3:\n")
        if gate =="u3":
            theta=float(input("Enter \u03B8:\n"))
        if gate in ["u3","u2","u1"]:
            phi=float(input("Enter \u03D5:\n"))
        if gate in ["u3","u2"]:
            lam=float(input("Enter \u03BB:\n"))
        if gate in ["u3","u2","u1"]:
            print("Building animation...")
            create_images(gate,theta,phi,lam)
            save_gif(gate)
        else:
            print("Not a valid gate, try again...")
            
if __name__ == '__main__':
    main()
```
